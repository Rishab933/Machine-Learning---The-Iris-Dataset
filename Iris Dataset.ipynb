# What is baseline modeling?
baseline modeling is just randomly gussing the species of the flower or gussing a single species 
for every data point we would expect to have a model accuracy of 33% since we have 3 different
classes that are evenly balanced

```python
# if we see the graph of petal width vs petal length 
# then we can clear justify the difference between the 3 flowers 
def single_feature(petal_length):
    if petal_length < 2.5:
        return 0
        #return "setosa"
    elif petal_length < 4.8:
        return 1
    else:
        return 2
```

```python
df_train.columns
```

```python
# 3th one is the petal length our conserned variable
```

```python
x_train[:,2]
# firstly x_train had all the values but i want to look at the petal length therefore
# calling x_train[:,2] -->[row,column] = [:(this includes all the rows), column no.]
```

```python
prediction_1 = np.array([single_feature(val) for val in x_train[:,2]])
```

```python
prediction_1
```

```python
prediction_1 == y_train
```

```python
# if the prediction that we have made is correct or it matches with the y_train which has the target values
# then it returns true otherwise false
# this prediction was made of the x_train data that too on a single variable as the petal length
```

```python
# now calculation the accuracy % of the prediction
acc = np.mean(prediction_1 == y_train)
```

```python
print(f"Accuracy of prediction_1 is: {acc*100:.2f}%")
```

```python
# its have a 95.5% accuracy of prediction_1 modeling or the single_feature modeling function
```

# USING CROSS VALIDATION TO EVALUATE OUR MODEL
In cross validation say suppose it divides the dataset in 5 group and say first group is for testing and other 4 group is for training
next the 2nd group is for testing and 1,3,4,5 is for training and so on

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
model = LogisticRegression(max_iter = 200) # for default the max iteration is 100 but we can change it to any number as this. 
```

```python
#cross_val_score?
# it would take estimator as the model,2 training datas as x & y, scoring, cv)
# expecting this 5 datas
# cv means number of folds
```

```python
cross_val_score(model, x_train,y_train,cv=5,scoring="accuracy")
```

```python
def plot_incorrect_prediction(df_prediction,x_axis_feature,y_axis_feature):
    fig,axs = plt.subplots(2,2,figsize=(10,10))
    axs = axs.flatten()
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature,hue="prediction_lable", data = df_prediction, ax = axs[0])
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature,hue="target_name", data = df_prediction, ax = axs[1])
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature,hue="correct_prediction", data = df_prediction, ax = axs[2])
    axs[3].set_visible(False) # to turn off the last grid
    plt.show()
```
